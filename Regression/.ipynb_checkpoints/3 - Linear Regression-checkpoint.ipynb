### load data

```python
import pandas as pd
from pylab import *
import numpy as np
df=pd.read_csv("Forex.csv",sep=",")
```

plot EUR-GBP time-serie

```python
x=df.index
y=df.EG
scatter(x,y)
```

### 1st-order linear regression

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

import matplotlib.pyplot as plt

def predict(o):
    return slope * o + intercept

fitLine = predict(x)

plt.scatter(x, y)
plt.plot(x, fitLine, c='r')
plt.show()
```

some statistics

```python
print("r2=%s" % r_value ** 2)
r, p = stats.pearsonr(y,fitLine)
print("corr=%s" % r)
```

### n-order linear regression

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(15,3))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

df.plot(x='time', y='EG', ax=ax1, legend=False)
df.plot(x='time', y='EU', ax=ax2, legend=False)
df.plot(x='time', y='GU', ax=ax3, legend=False)
ax1.set_title('EURGBP')
ax2.set_title('EURUSD')
ax3.set_title('GBPUSD')
```

```python
from sklearn.metrics import r2_score

fig, axs = plt.subplots(1,3, figsize=(15, 4))
axs = axs.ravel()

for i in range(3):
    p2 = np.poly1d(np.polyfit(x, y, i+2))
    axs[i].scatter(x, y)
    axs[i].plot(x, p2(x), c='r')
    axs[i].set_title("n = "+str(i+2))
    r2 = r2_score(y, p2(x))
    print("r2_"+str(i+2)+" = " + str(r2))
```

as we can see by increase the order or linear regression we get a closer fit.

### Train - Test

we will split the data into train (80%) and test (20%) and fit them with n=3 regression

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=.2, random_state=33)

p2 = np.poly1d(np.polyfit(xtrain, ytrain, 3))
```

```python
plt.scatter(xtrain, ytrain)
plt.scatter(xtrain, p2(xtrain), c='r')
```

```python
fig, axs = plt.subplots(1,2, figsize=(15, 4))
axs = axs.ravel()

axs[0].scatter(xtrain, ytrain)
axs[0].scatter(xtrain, p2(xtrain), c='r')
axs[0].set_title("train", fontsize=20)

axs[1].scatter(xtest, ytest)
axs[1].scatter(xtest, p2(xtest), c='r')
axs[1].set_title("test", fontsize=20)
```

```python
print("r2 train = %s" % r2_score(ytrain, p2(xtrain)))
print("r2 test = %s" % r2_score(ytest, p2(xtest)))
```
